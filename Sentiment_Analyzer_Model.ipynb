```python
def message_cleaning(message):
  punc_removed= [char for char in message if char not in string.punctuation]
  punc_removed_join= "".join(punc_removed) 
  punc_removed_join_clean =[word for word in punc_removed_join.split() if word.lower() not in stopwords.words('english')]
  return punc_removed_join_clean
```

```python
tweet_df_clean=df['SentimentText'].apply(message_cleaning)
tweet_df_clean[5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer= message_cleaning)
tweets_countvectorizer = CountVectorizer(analyzer=message_cleaning, dtype='uint8').fit_transform(df['SentimentText'])
```

```python
tweets_countvectorizer.shape
```

```python
x =tweets_countvectorizer
```

```python
y =df['Sentiment']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size= 0.10)
```

```python
#!pip install sklearn
from sklearn.naive_bayes import MultinomialNB 
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_test= NB_classifier.predict(X_test)
cm= confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot =True)
```

```python
print(classification_report(y_test , y_predict_test))
```

```python
from sklearn.metrics import f1_score, accuracy_score
pred =NB_classifier.predict(X_test)
print('f1_score is: '+ str(f1_score(y_test, pred)*100)+' %')
```

```python
print('Acccuracy of model is : '+ str(accuracy_score(y_test,pred)*100) +' %')
```
